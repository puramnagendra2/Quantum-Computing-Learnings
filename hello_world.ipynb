## Hello World Program

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

sampler = StatevectorSampler()
result = sampler.run([qc], shots=1024).result()
print(result)
```

```python
qc.draw()
```
